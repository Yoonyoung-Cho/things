## 전처리

__Ability__ 전처리

```python
%%time

# event = ability 인 경우 
# 이 경우에는 event_contents 에 위치정보가 있는 경우와 없는 경우로 나뉠 수 있어 
# 각각 ability_location과 ability_non_location으로 저장했다.
data = train
data_ability =data[data["event"]=="Ability"]

#위치정보가 포함된 경우
data_event_location = data_ability[data_ability["event_contents"].map(lambda x: x[x.find(":")])==":"]
event_specific=data_event_location["event_contents"].map(lambda x: x[8:x.find(";")])
ability_location = pd.concat([data_event_location.drop('event_contents',axis=1),event_specific],axis=1)

#위치정보가 포함되지 않은경우
data_event_non_location = data_ability[data_ability["event_contents"].map(lambda x: x[x.find(":")]) !=":"]
```

__Select__ 전처리

```python
%%time
#event = selection --> 세부 내용 저장
data_selection = train[train["event"]=="Selection"]
selection = data_selection["event_contents"].map(lambda x: x[2:x.find(" ")])
selection = pd.concat([data_selection.drop('event_contents',axis=1),selection],axis=1)
```

```python
temp = pd.concat([ability_location, data_event_non_location,selection],axis=0)
temp
# 자이제 event를 지워버리고 content를 event로 만들어서 
# ability나 selection이 아닌 나머지 train들과 합치기,,,,
```

```python
%%time

temp2 = temp.drop('event',axis=1)#.rename(columns = {'event_contents' : 'event'}, inplace = True)
temp2.columns = ['game_id', 'winner', 'time', 'player', 'species', 'event']
temp2 = pd.concat([temp2, train[(train["event"]!="Ability")&(train["event"]!="Selection")].drop('event_contents',axis=1)],axis=0)
temp2.head()
```
